```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

plt.style.use('seaborn')
sns.set(font_scale=2.5) 
import missingno as msno

#ignore warnings
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
df=pd.read_csv('drive/MyDrive/deeplearningforeveryone/dataset/pima-indians-diabetes.csv')
```

```python
df.info()
```

```python
df[['pregnant','class']].groupby(['pregnant'],as_index=True).mean().sort_values(by='pregnant',ascending=True)
```

```python
plt.figure(figsize=(12,12))
sns.heatmap(df.corr(), linewidths=0.1, vmax=0.5, cmap='summer_r', linecolor='white', annot=True)
plt.show()
```

```python
grid=sns.FacetGrid(df, col='class')
grid.map(plt.hist, 'plasma', bins=20)
plt.show()
```

```python
seed=1
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
X=df.iloc[:,0:8]
Y=df.iloc[:,8]
```

```python
model=Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X,Y, epochs=200, batch_size=40)
```
