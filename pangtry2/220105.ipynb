```python
import os
os.chdir('drive/MyDrive/pangdataset')

import pandas as pd
import numpy as np
from pycaret.regression import *
```

```python
train= pd.read_csv('train.csv')
test=pd.read_csv('test.csv')

train=train.drop('id', 1)
```

```python
train.head()
```

```python
setup(train, target='Body Mass (g)', session_id = 1234)
```

```python
compare_models(sort='RMSE')
```

```python
llar= create_model('llar', cross_validation = True)
lasso = create_model('lasso', cross_validation= True)
lr= create_model('lr', cross_validation = True)
```

```python
tune_llar= tune_model(llar, optimize='RMSE', n_iter = 1000)
tune_lasso= tune_model(lasso, optimize='RMSE', n_iter = 1000)
tune_lr= tune_model(lr, optimize='RMSE', n_iter = 1000)
```

```python
blended_specifier = blend_models(estimator_list = [tune_llar,tune_lasso, tune_lr ], fold=10, optimize='RMSE')
```

```python
final_model=finalize_model(blended_specifier)
```

```python
submission=pd.read_csv('sample_submission.csv', index_col=0)
```

```python
submission.head()
```

```python
test=test.drop('id',1)
test.head()
```

```python
pred=predict_model(final_model, data=test)
pred.head()
```

```python
submission['Body Mass (g)']=pred['Label']
submission.head()
```

```python
submission.to_csv('pycraret2.csv')
```
